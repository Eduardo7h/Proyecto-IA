```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Importamos el csv
df = pd.read_csv('/content/drive/MyDrive/DataSet/spotify_tracks_dataset.csv')
#modificar ruta /content/drive/MyDrive/DataSet/spotify_tracks_dataset.csv
# Creamos una copia del data
cp_df = df.copy()
cp_df
```

```python
#Regresión
#X es la energía de la canción y queremos prodecir el volumen (loudness)

X= cp_df [['energy']]
y= cp_df ['loudness']

#regresión lineal
modelo_spotify= LinearRegression()
#entrenar
modelo_spotify.fit(X,y)

print("Modelo de regresión")
```

```python
#sacar la pendiente y=mx
pendiente = modelo_spotify.coef_[0]
interseccion= modelo_spotify.intercept_
print (f"Si energy sube, el loudness aumenta: {pendiente:.2f} decibeles.")
print (f"La ecuación de la pendiente es: Loudness= ({pendiente:.2f} *Energy) + {interseccion:.2f} ")
```

```python
predY=modelo_spotify.predict(cp_df[['energy']])
plt.plot(cp_df.energy.values,predY,c='red',linestyle='-',linewidth=0.5)
plt.scatter(cp_df['energy'], cp_df['loudness'], alpha=0.5, label='Datos reales')
plt.xlabel("Energy (calm/ energetic)")
plt.ylabel("Loudness (dB)")
plt.title("Regression")
plt.show()
```

```python
#Ejemplo de una canción
new_energy= pd.DataFrame({'energy'== [0.5]})
prediccion_volumen= modelo_spotify.predict(new_energy)

print(f"La nueva canción con 0.5 de energía, calculamos que tendrá loudness de {prediccion_volumen[0]:.2f} dB")
```

```python
fig, axs = plt.subplots(1, 1, figsize=(5, 5))
sns.kdeplot(x='energy', y='loudness', data=cp_df, ax=axs, color='#065A60')
axs.set_title('Energy vs loudness')
axs.grid()
plt.tight_layout()
plt.show()
```

```python
#Clasificación
#Predecir si una canción es feliz o triste (Mode= 0/1) basandose en danceability, energy y speechiness.
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#Separar los datos
X_clas = cp_df[['danceability', 'energy', 'speechiness', 'liveness', 'tempo', 'instrumentalness', 'acousticness', 'duration_ms' ]].dropna()

#Definir si una canción es feliz (1) o triste (0)
y_clas = cp_df['mode'].dropna()

# Dividimos los datos: 70% para que el modelo estudie, 30% para hacerle un examen final.
X_train, X_test, y_train, y_test = train_test_split(X_clas, y_clas, test_size=0.2, random_state=42)

# Creamos nuestro modelo con Random Forest.
#modelo_clasificacion = RandomForestClassifier(random_state=42)

# Lo ponemos a entrenar
#modelo_clasificacion.fit(X_train, y_train)

#print("Listo")
```

```python
modelo_premium = RandomForestClassifier(
    n_estimators=300,        # subimos árboles para que piense mejor la respuesta
    max_depth=30,            # Le ponemos un límite para evitar el sobreajuste
    random_state=42
)

# Lo ponemos a entrenar con las nuevas configuraciones
modelo_premium.fit(X_train, y_train)

# Que haga el examen final
predicciones_premium = modelo_premium.predict(X_test)
calificacion_nueva = accuracy_score(y_test, predicciones_premium)

print(f"¡Momento de la verdad! Nuestra nueva calificación es: {calificacion_nueva * 100:.2f}%")
```
